### Load 2000 SIZ

```python
dir_in = '/glade/campaign/cgd/ppc/duvivier/cesm2_antarctic_polynya/mpa_analysis/DATA/ecoindex_data/sea_ice/'
```

```python
# Read in 2000 data
file_in = 'CESM2-LE-SIZ.nc'
ds_siz = xr.open_mfdataset(dir_in+file_in,decode_times=False)

# get ensemble mean SIC for september in 2000 and 2090
sic_2000 = ds_siz.SIC_SEP.mean(dim='member_id').sel(time=2000)
sic_2090 = ds_siz.SIC_SEP.mean(dim='member_id').sel(time=2090)
# make array to fill
siz_2000 = sic_2000.where(sic_2000 == 1, other = 1)
siz_2090 = sic_2090.where(sic_2090 == 1, other = 1)
# keep only SIZ data
siz_2000 = siz_2000.where(sic_2000 >= 0.15)
siz_2090 = siz_2090.where(sic_2090 >= 0.15)
```

```python
siz_2000.plot()
```

```python
siz_2090.plot()
```

```python
# make mask that shows only places where SIZ was LOST by 2090
# first convert siz arrays to have values of 2 in ocean but 1 over sea ice (ignore land, it doesn't change)
siz_2000_2 = siz_2000.fillna(2)
siz_2090_2 = siz_2090.fillna(2)

siz_2000_2.plot()
#siz_2090_2.plot()
```

```python
# Calculate differences
# Areas where both decades had ocean should be zero
# Areas where both decades had ice should be zero
# Areas with 1 are where it used to be ice in 2000 and is now ocean in 2090
temp = siz_2090_2 - siz_2000_2

temp.plot()
```

```python
# Make a SIZ change mask

# make array to fill
siz_diff = sic_2000.where(sic_2000 == 1, other = 1)
# now keep only points that are 1 above 
siz_diff = siz_diff.where(temp >= 0.5)

siz_diff.plot()
```

## Calculate AEV difference by decades

```python
# calculate difference between 2090s and 2000s for AEV
data_2000 = ds_eco.ecoindex_regional_decadal.sel(decades=2000)
data_2090 = ds_eco.ecoindex_regional_decadal.sel(decades=2090)
diff = data_2090 - data_2000

diff.plot()
```

```python
# note that many of the blue places indicate the SIZ has shrunk, so we keep only areas where the SIZ exists in 2090
diff_masked = diff * siz_2090

diff_masked.plot()
```
