## Load FESITY data from CESM

```python
path_in = '/glade/campaign/cgd/ppc/duvivier/cesm2_antarctic_polynya/mpa_analysis/DATA/ecoindex_data/marbl/'
```

```python
# load hindcast data
file_in = 'CESM2-JRA55-hindcast-annual-mean-FEISTY.nc'
ds_in = xr.open_dataset(path_in+file_in, decode_times=True)

# keep demersal fish
df_hindcast = ds_in.Demersal
```

```python
# load cesm2-le data
file_in = 'CESM2-LE-annual-mean-FEISTY.nc'
ds_in = xr.open_dataset(path_in+file_in, decode_times=True)

# keep forage fish
df_cesm2le = ds_in.Demersal
```

```python
# trim both datasets to 2003-2020
df_hindcast = df_hindcast.isel(time=slice(42,63))
df_cesm2le  = df_cesm2le.isel(time=slice(50,71))
```

```python
# take averages over all years and all ensemble members (for cesm2-le)
df_hindcast_mean = df_hindcast.mean(dim=('time'))
df_cesm2le_mean = df_cesm2le.mean(dim=('time','member_id'))
```

## Plot figure

```python
# make circular boundary for polar stereographic circular plots
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
```

```python
# get colony distance info for plotting
lons = ds_grid.TLONG
lats = ds_grid.TLAT
```

```python
# mask to omit land points
df_hindcast_mean = df_hindcast_mean.where(ds_grid.REGION_MASK == 1)
df_cesm2le_mean = df_cesm2le_mean.where(ds_grid.REGION_MASK == 1)
```

### Set plotting ranges, etc.

```python
# set some plotting info
cmap_in = cmocean.cm.turbid
vmin_in = 0
vmax_in = 6.0
units = 'g WW/m^2'
```

### Get shapefile with geography

```python
dir_in = '/glade/campaign/cgd/ppc/duvivier/cesm2_antarctic_polynya/mpa_analysis/DATA/'
```

```python
# Read in Antarctic Geography Shapefiles
file_in = 'USNIC_ANTARC_shelf_2022/USNIC_ANTARC_shelf_2022.shp'

shdf_ant = gpd.read_file(dir_in+file_in)
# Important: convert the projection from the shapefile one to the cartopy one 
shdf_ant = shdf_ant.to_crs(ccrs.SouthPolarStereo())
```

### Create a custom colormap for the Aquamap

```python
from matplotlib import colors as c
cMap_0 = c.ListedColormap(['khaki'])
cMap_80 = c.ListedColormap(['darkgoldenrod'])

# look at colormap
#data = np.random.rand(10, 10) # A 10x10 array of random values
#plt.imshow(data, cmap=cMap_80)
```

### Save plot

```python
# set path for figure output
dir_out = '/glade/campaign/cgd/ppc/duvivier/cesm2_antarctic_polynya/mpa_analysis/ecoindex_figures/version_6/'
fout = 'SFigure_FEISTY_Demersal_validation'
```

```python
# Plot
# create figure
fig = plt.figure(figsize=(20,20))

# plot the region as subplots - note it's nrow x ncol x index (starting upper left)

### Plot Obs ###
ax = fig.add_subplot(1,3,1, projection = ccrs.SouthPolarStereo())
ax.set_boundary(circle, transform=ax.transAxes)
ax.set_extent([0.005, 360, -90, -60], crs=ccrs.PlateCarree())
ax.set_title('a) Aquamaps Toothfish Habitat', loc='left')
# 0% occurrance
this=ax.pcolormesh(ds_aquamaps_toothfish_0.lon,ds_aquamaps_toothfish_0.lat,
                     ds_aquamaps_toothfish_0.presence.where(ds_aquamaps_toothfish_0.presence==1),
                     cmap=cMap_0,                      
                     transform=ccrs.PlateCarree())
# 80% occurrance
this=ax.pcolormesh(ds_aquamaps_toothfish_80.lon,ds_aquamaps_toothfish_80.lat,
                     ds_aquamaps_toothfish_80.presence.where(ds_aquamaps_toothfish_80.presence==1),
                     cmap=cMap_80,                      
                     transform=ccrs.PlateCarree())

# Add shapefile coastlines
ax.add_geometries(shdf_ant.geometry , ccrs.SouthPolarStereo(), facecolor='none', edgecolor='black',linewidth=1, zorder=10)
# Add legend
this = ax.text(340.0,-81.0,'Likelihood\nof occurrence',color='black', fontsize='large',transform=ccrs.PlateCarree())
this = ax.scatter(0.0,-84.5, color='darkgoldenrod', edgecolor='black',s=250,transform=ccrs.PlateCarree())
this = ax.text(30.0,-85.0,'>80%',color='black', fontsize='large',transform=ccrs.PlateCarree())
this = ax.scatter(0.0,-89.0, color='khaki', edgecolor='black',s=250,transform=ccrs.PlateCarree())
this = ax.text(90.0,-87,'>0%',color='black', fontsize='large',transform=ccrs.PlateCarree())

### Plot CESM2 hindcast ###
ax = fig.add_subplot(1,3,2, projection = ccrs.SouthPolarStereo())
ax.set_boundary(circle, transform=ax.transAxes)
ax.set_extent([0.005, 360, -90, -60], crs=ccrs.PlateCarree())
ax.set_title('b) Demersal Fish from CESM2-HistRecon', loc='left')
lon, lat, field = utils.adjust_pop_grid(lons,lats,df_hindcast_mean)
this=ax.pcolormesh(lon,lat,
                     field,
                     cmap=cmap_in,                      
                     vmax=vmax_in, vmin=vmin_in,
                     transform=ccrs.PlateCarree())
# Add shapefile coastlines
ax.add_geometries(shdf_ant.geometry , ccrs.SouthPolarStereo(), facecolor='none', edgecolor='black',linewidth=1, zorder=10)

### Plot CESM2-LE ###
ax = fig.add_subplot(1,3,3, projection = ccrs.SouthPolarStereo())
ax.set_boundary(circle, transform=ax.transAxes)
ax.set_extent([0.005, 360, -90, -60], crs=ccrs.PlateCarree())
ax.set_title('c) Demersal Fish from CESM2-LE', loc='left')
lon, lat, field = utils.adjust_pop_grid(lons,lats,df_cesm2le_mean)
this=ax.pcolormesh(lon,lat,
                     field,
                     cmap=cmap_in,                      
                     vmax=vmax_in, vmin=vmin_in,
                     transform=ccrs.PlateCarree())
# Add shapefile coastlines
ax.add_geometries(shdf_ant.geometry , ccrs.SouthPolarStereo(), facecolor='none', edgecolor='black',linewidth=1, zorder=10)

# Add a common colorbar
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.4, 0.02, 0.2])
cbar1 = fig.colorbar(this, cax=cbar_ax,extend='max',label='Demersal Fish Potential Biomass ('+units+')')

# Finalize figure and save
fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)
fig = plt.savefig(dir_out+fout+'.png', bbox_inches='tight', dpi=200)
```
