```python
# keep the variable of interest only
data = ds_in.TEMP_bottom
```

```python
# keep the bottom depth because we need to mask by the continental shelf
# all members and years have the same field here, so just grab member 1, time 1
depth = ds_in.HT.isel(year=0,member_id=0)
# convert to meters from cm
depth = depth/100.0
```

```python
# mask temperature to be continental shelf only
data = data.where(depth <= 2000.0)
```

```python
# reassign nlat and nlon for later
data['nlat'] = ds_grid.nlat
data['nlon'] = ds_grid.nlon
```

### Calculate TBOT decadal averages

```python
# specify some arrays that will become coordinate arrays
decades = list(range(1960,2100,10))
decade_names = ['1960','1970','1980','1990','2000','2010','2020','2030','2040','2050','2060','2070','2080','2090']
```

```python
# get dimensions of the grid we need to iterate over
xlen = len(ds_in.nlon)
ylen = len(ds_in.nlat)
dlen = len(decades)
print(xlen,ylen,dlen)
```

```python
# Loop through decades

# make numpy array to fill and specify dimensions
array_decadal_all = np.zeros([dlen,ylen,xlen])

for dec in decades:
    print('Calculating decadal means for '+str(dec)+'s')
    # need to specify which spot in the array we'll fill for the decade
    if dec == 1960:
        count = 0
    else: 
        count = count+1    
    # set the start and end years for this decade
    yy_st = dec - 2
    yy_ed = dec +3
    yy = list(range(yy_st,yy_ed,1)) 
    # get data subset
    temp1 = data.where((data.year >= yy_st)&(data.year < yy_ed),drop=True)
    # calculate mean over time
    temp2 = temp1.mean(dim='year')
    # get ensemble mean
    array_decadal_all[count,:,:] = temp2.mean(dim='member_id')

# convert to xarray
tbot_dec = xr.DataArray(array_decadal_all,dims=('decades','nlat','nlon'))
tbot_dec['decades'] = decades
tbot_dec['nlat'] = ds_grid.nlat
tbot_dec['nlon'] = ds_grid.nlon
```

```python
# save just two decades
tbot_2000 = tbot_dec.sel(decades=2000)
tbot_2090 = tbot_dec.sel(decades=2090)
```

```python
tbot_2090.plot()
```

```python
#create masks where the TBOT is less than 1C threshold for Toothfish
tbot_mask_2000 = mask_ones.where(tbot_2000 <=1.0)
tbot_mask_2090 = mask_ones.where(tbot_2090 <=1.0)
```
